# # Task 1: Summarization

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("t5-small")
model = AutoModelForSeq2SeqLM.from_pretrained("t5-small")
```

```python
text = """
Machine learning is a subset of artificial intelligence that focuses on building
systems that learn from data. These systems improve their performance over time
without being explicitly programmed.
"""
```

```python
inputs = tokenizer(
    "summarize: " + text,
    return_tensors="pt",
    truncation=True
)
```

```python
outputs = model.generate(
    **inputs,
    max_new_tokens=10,
    
)
```

```python
decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(f"Summarize output: {decoded}")
```

# Task 2: Translation (Same Model!)

```python
inputs = tokenizer("translate English to German: The house is wonderful.", return_tensors="pt")
outputs = model.generate(**inputs)


decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(f"Translation output: {decoded}")
```

# Task 3 - Question Answering

```python
context = """
Natural Language Processing is a field of artificial intelligence
that focuses on enabling computers to understand human language.
"""

question = "What is focuses on?"
input_text = f"question: {question} context: {context}"

inputs = tokenizer(input_text, return_tensors="pt")
outputs = model.generate(**inputs)


decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(f"Question Answering output: {decoded}")
```
